```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#设置显示列数
pd.set_option('display.max_columns', None)
#显示所有行
pd.set_option('display.max_rows', None)
#设置value的显示长度为100，默认为50
pd.set_option('max_colwidth',100)
```

```python
df = pd.read_csv('./used_car_train_20200313.csv', sep=' ')
```

```python
# nan可视化
missing = df.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar(rot=0,color=['r','g','b'])
```

```python
# 特征nunique分布
cat_fea = df.columns
for fea in cat_fea:
    print("{}特征有个{}不同的值".format(fea,df[fea].nunique()))
    #print(df[fea].value_counts())
```

# 类型转换

```python
df["notRepairedDamage"].unique()
```

```python
# 替换值
df["notRepairedDamage"] = df["notRepairedDamage"].replace('-',np.NaN)
df["notRepairedDamage"].unique()
```

```python
# 类型转换
df["notRepairedDamage"] = df["notRepairedDamage"].astype("float")
```

```python
df["notRepairedDamage"].unique() 
```

# 数据信息展示

```python
# 画直方图
plt.figure()
plt.figure(figsize=(16, 10))
i = 1
for fea in df.columns:
    if df[fea].nunique()<20:
        plt.subplot(2,4, i)
        i += 1
        v = df[fea].value_counts()
        fig = sns.barplot(x=v.index, y=v.values)
        for item in fig.get_xticklabels():
            item.set_rotation(45)
        plt.title(fea)
plt.tight_layout()
plt.show()
```

```python
# 特征可视化
num_fea = df.columns
f = pd.melt(df, value_vars=num_fea)
g = sns.FacetGrid(f, col="variable",col_wrap=3, sharex=False, sharey=False)
g = g.map(sns.distplot, "value")
```
