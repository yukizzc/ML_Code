```python
data = pd.read_csv('./fttt.csv',encoding='gbk')
```

```python
y = data['happiness']
data2 = data.drop(['id','happiness'],axis=1)
x = data2
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
clf = RandomForestRegressor()
scores = cross_val_score(clf, x.values, y.values, cv=5, scoring='neg_mean_squared_error')
```
