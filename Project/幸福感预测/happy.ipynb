```python
# nan可视化
plt.figure(figsize=(16, 10))
missing = data.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar(rot=0,color=['r','g','b'])
plt.xticks(color='blue',rotation=45)
plt.show()
```
